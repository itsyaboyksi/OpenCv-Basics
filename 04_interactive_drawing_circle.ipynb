```python
# Import Library
import cv2

# Callback Function for the Mouse, Circle
def draw_circle(event, x, y, flags, param):
    
    
    # write your global params.  
    # Check the Project Computer Vision - Image Basics with OpenCV and Python
    # to get an idea how you can do it
    global pt1, clicked 
    
    # Create an event for Left Button Down and assigned it to event
    if event == cv2.EVENT_LBUTTONDOWN:

    # Reset your Circle
        if clicked == True:
        # Give 0 values to pt1 
            pt1 = (0,0)
        # Give False value to your Trackers
            clicked = False
            
    # Check the top_left_clicked if it's False
    if clicked == False:
        pt1 = (x, y)
        clicked = True 
    
    
# pt1 is a Global Variable
pt1 = (0,0)


# Tracker is False
clicked = False

# Take a video
cap = cv2.VideoCapture(0)


# Create a Named Window for the Connections
cv2.namedWindow('Test')


# Set a Mouse Callback
cv2.setMouseCallback('Test', draw_circle)

# Time for Magic
while True:
    
    # Capture the frame
    ret, frame = cap.read()
    
    # Based on the Global Variables Draw the Frame
    # Draw a Circle on the Frame
    if clicked == True:
        cv2.circle(frame, center = pt1, radius = 55, color = (255,0,255), thickness = -1) #setting thickness to -1 will make a filled circle
    
    # Show the Frame
    cv2.imshow('Test', frame)
    
    # 27 == esc button. You can use any other button like ord('q')
    # If you try to close the window pressing the x you'll get in trouble (^_^)
    if cv2.waitKey(15) & 0xFF == 27:
        break
    
    
# Never forget first to release
# And then to destroy   

cap.release()
cv2.destroyAllWindows()

```
